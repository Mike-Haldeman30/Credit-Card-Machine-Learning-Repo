```python
train_df.drop(columns=['SSN', 'Name'], axis=1, inplace=True)
```

```python
categorical_cols = [c for c in train_df.columns if train_df[c].dtype == 'object']


for col in categorical_cols:
    train_df[col] = train_df[col].str.strip('_')

    try:
        train_df[col] = train_df[col].astype('float64')
    except:
        train_df[col] = train_df[col]
        

for col in categorical_cols:
    train_df[col] = train_df[col].replace({'':np.nan})

    try:
        train_df[col] = train_df[col].astype('float64')
    except:
        train_df[col] = train_df[col]

for col in categorical_cols:
    train_df[col] = train_df[col].replace({'!@9#%8':np.nan, '#F%$D@*&8':np.nan})
```

```python
train_df.head()
```

```python
train_df['Credit_History_Age']
#just taking year from this data 

credit= []

for i in train_df['Credit_History_Age']:
     credit.append(str(i).split(' ')[0])
        

train_df['Credit_History_Age'] = credit




#need to make null values an actual value rather than a string 
train_df['Credit_History_Age'] = train_df['Credit_History_Age'].replace({'nan':np.nan})
train_df['Credit_History_Age'] = train_df['Credit_History_Age'].astype('float64')




train_df.head()
```

```python
#This is used to remove outliers within the dataset, We can see above there is a scientist that is -500 for age


numerical_cols = [col for col in train_df.columns if (train_df[col].dtype == 'int64') | (train_df[col].dtype == 'float64')]

for x in list(numerical_cols):
    q75,q25 = np.percentile(train_df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    train_df.loc[train_df[x] < min,x] = np.nan
    train_df.loc[train_df[x] > max,x] = np.nan
    
train_df.head()
```
